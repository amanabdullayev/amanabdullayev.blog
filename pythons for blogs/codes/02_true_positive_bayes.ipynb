```python
import warnings

warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import graphviz as gr
from matplotlib import style
import seaborn as sns
from matplotlib import pyplot as plt

plt.rcParams["figure.dpi"] = 100
%config InlineBackend.figure_format = "retina"
style.use("seaborn-v0_8-colorblind")

%load_ext jupyter_black
```

```python
plt.xkcd()
g = gr.Digraph()

g.edge(
    "Person",
    "Sick",
    label="P(sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.edge(
    "Person",
    "Not Sick",
    label="P(not sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.edge(
    "Sick",
    "Tested Positive \n(True Positive)",
    label="P(positive|sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.edge(
    "Sick",
    "Tested Negative \n(False Negative)",
    label="P(negative|sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.edge(
    "Not Sick",
    "Tested Positive \n(False Positive)",
    label="P(positive|not sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.edge(
    "Not Sick",
    "Tested Negative \n(True Negative)",
    label="P(begative|not sick)",
    fontsize="10",
    fontcolor="blue",
    fontname="helvetica",
    fontweight="bold",
)
g.node("Person", style="filled", fillcolor="#00677f", shape="box")
g.node("Sick", style="filled", fillcolor="#ff6e54")
g.node("Not Sick", style="filled", fillcolor="#94dc7b")
g.node("Tested Positive \n(True Positive)", style="filled", fillcolor="#dd5182")
g.node("Tested Negative \n(False Negative)", style="filled", fillcolor="#ffa600")
g.node("Tested Positive \n(False Positive)", style="filled", fillcolor="#f9f871")
g.node("Tested Negative \n(True Negative)", style="filled", fillcolor="#2db88b")
# fill nodes with light blue
# save image as png
g.render("images/bayes_net", format="png", cleanup=True)
g
```
